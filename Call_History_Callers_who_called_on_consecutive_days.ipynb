```python
import pandas as pd
import sqlite3

# Sample caller_history data
data = {
    "caller_id": [1, 1, 1, 2, 2, 3, 3, 3],
    "date": [
        "2024-01-01",
        "2024-01-02",
        "2024-01-04",
        "2024-02-10",
        "2024-02-11",
        "2024-03-01",
        "2024-03-03",
        "2024-03-04",
    ]
}

df = pd.DataFrame(data)
df["date"] = pd.to_datetime(df["date"])

# Create in-memory SQLite database
conn = sqlite3.connect(":memory:")
df.to_sql("caller_history", conn, index=False, if_exists="replace")

# SQL: find callers who called on consecutive days
query = """
WITH cte AS (
    SELECT
        caller_id,
        date,
        LAG(date) OVER (
            PARTITION BY caller_id
            ORDER BY date
        ) AS prev_date
    FROM caller_history
)
SELECT DISTINCT
    caller_id
FROM cte
WHERE julianday(date) - julianday(prev_date) = 1;
"""
#Use DATEDIFF(day, prev_date, date) for SQL Server

result = pd.read_sql(query, conn)

print("Caller History:")
print(df)

print("\nCallers with consecutive-day calls:")
print(result)
```
