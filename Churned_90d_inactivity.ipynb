```python
import sqlite3
from datetime import date, timedelta

# Create in-memory SQLite database
conn = sqlite3.connect(":memory:")
cursor = conn.cursor()

# Create table
cursor.execute("""
CREATE TABLE fact_sales (
    user_id INTEGER,
    dt DATE
)
""")

today = date.today()

# Sample Data
# User 1 → Active recently (NOT churned)
# User 2 → Last order 120 days ago (CHURNED)
# User 3 → Last order 200 days ago (CHURNED)
# User 4 → Active 10 days ago (NOT churned)

sample_data = [
    (1, today - timedelta(days=10)),
    (1, today - timedelta(days=30)),

    (2, today - timedelta(days=120)),

    (3, today - timedelta(days=200)),

    (4, today - timedelta(days=10)),
]

cursor.executemany("INSERT INTO fact_sales VALUES (?, ?)", sample_data)
conn.commit()

# Churn Query (SQLite syntax adjustment)
query = """
SELECT DISTINCT f1.user_id
FROM fact_sales f1
LEFT JOIN fact_sales f2
  ON f1.user_id = f2.user_id
  AND f2.dt >= date('now', '-90 day')
WHERE f2.user_id IS NULL;
"""

cursor.execute(query)
results = cursor.fetchall()

print("Churned Users:", results)

conn.close()
```
