# Determine 7-day retention: users active at least 3 days in the last week

```python
import sqlite3
from datetime import date, timedelta

# Create in-memory database
conn = sqlite3.connect(":memory:")
cursor = conn.cursor()

# Create table
cursor.execute("""
CREATE TABLE fact_sales (
    user_id INTEGER,
    dt DATE
)
""")

# Insert sample data
today = date.today()

sample_data = [
    (1, today - timedelta(days=1)),
    (1, today - timedelta(days=2)),
    (1, today - timedelta(days=3)),  # user 1 active 3 days

    (2, today - timedelta(days=1)),
    (2, today - timedelta(days=4)),  # user 2 active 2 days

    (3, today - timedelta(days=1)),
    (3, today - timedelta(days=2)),
    (3, today - timedelta(days=3)),
    (3, today - timedelta(days=4)),  # user 3 active 4 days
]

cursor.executemany("INSERT INTO fact_sales VALUES (?, ?)", sample_data)
conn.commit()

# 7-day retention query (SQLite syntax)
query = """
WITH last_week_activity AS (
    SELECT
        user_id,
        COUNT(DISTINCT dt) AS active_days
    FROM fact_sales
    WHERE dt >= date('now', '-7 day')
    GROUP BY user_id
)
SELECT user_id
FROM last_week_activity
WHERE active_days >= 3;
"""

cursor.execute(query)
results = cursor.fetchall()

print("Retained users:", results)

conn.close()
```
